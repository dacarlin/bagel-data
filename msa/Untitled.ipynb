```python
from Bio import AlignIO
from collections import Counter 

alignment = AlignIO.read( 'msa_aligned_to_bagel.fasta', 'fasta' )
```

```python
pos120 = alignment.get_column(120 - 1)
Counter( pos120 ) 
```

```python
ws = []
fs = []
ys = []
i = 120 

for record in alignment:
    if record.seq[ i - 1 ] == 'w':
        ws.append( record.name ) 
    elif record.seq[ i - 1 ] == 'f':
        fs.append( record.name ) 
    elif record.seq[ i - 1 ] == 'y':
        ys.append( record.name ) 
```

```python
import requests 

for w in ws[1:3]: 
    r = requests.get( 'http://www.uniprot.org/uniprot/?query={}&columns=id,organism-id&format=tab'.format( w.split( '_' )[0] ) ) 
    print r.text
```

```python
from ete2 import Tree, TreeStyle, NodeStyle 

t = Tree( "clustalo-I20160308-053834-0263-7021484-es.ph.txt" )
```

```python
# big red dot 
big_red = NodeStyle()
big_red["hz_line_color"] = "red"
big_red["size"] = 50545

# apply to Bagel 
node = t.search_nodes(name=u"BGLB_PAEPO__modified_")[0]
node.set_style( big_red ) 

# traverse and color 
for n in t.traverse():
    if n.name in fs:
        nstyle = NodeStyle()
        nstyle["hz_line_color"] = "RosyBrown"
        n.set_style( nstyle )
    elif n.name in ws:
        nstyle = NodeStyle()
        nstyle["hz_line_color"] = "SandyBrown"
        n.set_style( nstyle )
    elif n.name in ys:
        nstyle = NodeStyle()
        nstyle["hz_line_color"] = "#f4ee60"
        n.set_style( nstyle )

# write tree 
circular_style = TreeStyle()
circular_style.mode = "c" # draw tree in circular mode
t.render( "tree.pdf", w=20000, h=20000, tree_style=circular_style )
```
